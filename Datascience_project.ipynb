```python
import sys
!{sys.executable} -m pip install yfinance
import yfinance as yf

tesla_data = yf.download('TSLA')
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import pandas as pd

tesla_revenue = pd.DataFrame({
    'Date': ['2023-12-31', '2024-03-31', '2024-06-30', '2024-09-30', '2024-12-31'],
    'Revenue': [25000, 27000, 29000, 31000, 33000]
})
tesla_revenue.tail()
```

```python
gme_data = yf.download('GME')
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
gme_revenue = pd.DataFrame({
    'Date': ['2023-12-31', '2024-03-31', '2024-06-30', '2024-09-30', '2024-12-31'],
    'Revenue': [1500, 1700, 1900, 2100, 2300]
})
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

make_graph(tesla_data, 'Tesla Stock Price')
```

```python
make_graph(gme_data, 'GameStop Stock Price')
```
